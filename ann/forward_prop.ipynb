```python
import numpy as np
import matplotlib.pyplot as plt
```

Create 3 gaussian clouds

```python
N = 500
X1 = np.random.randn(N, 2) + np.array([0, -2])
X2 = np.random.randn(N, 2) + np.array([2, 2])
X3 = np.random.randn(N, 2) + np.array([-2, 2])
X = np.vstack([X1, X2, X3])
Y = np.array([0]*N + [1]*N + [2]*N)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, s=100, alpha=0.5)
```

Randomly initialize weights

```python
D = 2  # dimensionality of features
M = 3  # Hidden layer size
K = 3  # number of output classes
W1 = np.random.randn(D, M)
b1 = np.random.randn(M)
W2 = np.random.randn(M, K)
b2 = np.random.randn(K)
```

```python
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
```

```python
def forward(X, W1, b1, W2, b2):
    Z = sigmoid(X.dot(W1) + b1)
    A = Z.dot(W2) + b2
    expA = np.exp(A)
    Y = expA / expA.sum(axis=1, keepdims=True)
    return Y
```

```python
def classification_rate(Y, P):
    n_correct = 0
    n_total = 0
    for i in range(len(Y)):
        n_total +=  1
        if Y[i] == P[i]:
            n_correct += 1
    return float(n_correct) / n_total
```

```python
P_Y_given_X = forward(X, W1, b1, W2, b2)
P = np.argmax(P_Y_given_X, axis=1)
```

```python
print("Classification rate", classification_rate(Y, P))
```

```python
assert len(P) == len(Y)
np.sum(P == Y) / len(Y)
```
